```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/sample_data/homeprices.csv')
df
```

```python
df.area.shape
```

```python
plt.xlabel('area')
plt.ylabel('price')
plt.scatter(df.area,df.price)
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['area']],df.price)
```

```python
reg.predict([[5000]])
```

```python
pred_df=pd.read_csv('/content/sample_data/areas.csv')
```

```python
pred_df['price']=reg.predict(pred_df)
```

```python
pred_df
```
