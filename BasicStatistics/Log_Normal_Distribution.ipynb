```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
df = pd.read_csv("/content/sample_data/income.csv")
df
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x=df['income($)'],y=df['count'],data=df)
g.set_xticklabels(g.get_xticklabels(),
                          rotation=45,
                          horizontalalignment='right');
```

```python
sns.set(rc={'figure.figsize':(11.7,8.27)})
g = sns.barplot(x='income($)',y='count',data=df)
g.set_xticklabels(g.get_xticklabels(),
                          rotation=45,
                          horizontalalignment='right');
g.set(xscale="log");
```
