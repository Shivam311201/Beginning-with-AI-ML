```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/sample_data/AB_NYC_2019.csv")
df
```

```python
df["price"].describe()
```

```python
# Quantile helps in calculating percentile less than 0 in this case
df["price"].quantile(0)
```

```python
Q1 = df["price"].quantile(0.25)
```

```python
df["price"].quantile(0.5)
```

```python
Q3 = df["price"].quantile(0.75)
IQR=Q3-Q1
```

```python
df["price"].quantile(1)
```

```python
percentile_1,percentile_99 = df["price"].quantile([0.01, 0.99])
percentile_1,percentile_99
```

```python
Q1=max(30,Q1-1.5*IQR)
Q3=Q3+1.5*IQR
```

```python
IQR
```

```python
Q1,Q3
```

```python
df=df[(df["price"]<percentile_99)&(df["price"]>percentile_1)]
df["price"].describe()
```

```python
df=df[(df["price"]<Q3)&(df["price"]>Q1)]
df["price"].describe()
```

# Outliers in Height-Weight Dataset

```python
df=pd.read_csv("/content/sample_data/weight-height.csv")
df
```

```python
df["Height"].describe()
Q1H=df["Height"].quantile(0.25)
Q3H=df["Height"].quantile(0.75)
IQRH=Q3H-Q1H
```

```python
Q1H=Q1H-1.5*IQRH
Q3H=Q3H+1.5*IQRH
```

```python
df["Weight"].describe()
Q1W=df["Weight"].quantile(0.25)
Q3W=df["Weight"].quantile(0.75)
IQRW=Q3W-Q1W
```

```python
df=df[(df['Weight']>Q1W)&(df['Weight']<Q3W)&(df['Height']>Q1H)&(df['Height']<Q3H)]
df
```
